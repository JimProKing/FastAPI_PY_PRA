# 데코레이터 연습

```python
def temp_deco(func):
    def wrapper():
        print("함수호출 전")
        func()
        print("함수호출 후")
    return wrapper
```

```python
def hello_world():
    print("Hello, World!")

hello_world()
```

```python
pra = temp_deco(hello_world)

pra()
```

```python
def say_hi(name: str | None):
    print(f"Hey {name}!")

say_hi("minsu")
```
